```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv(r"C:\Users\navam\Downloads\archive (1)\Sample - Superstore.csv", encoding='latin1')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df['Month'] = df['Order Date'].dt.month_name()
```

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
total_profit = df['Profit'].sum()
print("Total Profit:", total_profit)
```

```python
total_orders = df['Order ID'].nunique()
print("Total Orders:", total_orders)
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='bar', figsize=(10,5))

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

```python
plt.plot(monthly_sales)
plt.title("Monthly Sales Trend")

plt.savefig('../visuals/monthly_sales.png')  # SAVE FIRST
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='barh', figsize=(10,6))

plt.title('Top Selling Products')
plt.xlabel('Sales')

plt.show()
plt.savefig('../visuals/top_products.png')
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))

plt.title('Region-wise Sales')
plt.ylabel('')

plt.show()
plt.savefig('../visuals/region_sales.png')
```

```python
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='bar', figsize=(8,5))

plt.title('Category-wise Profit')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.show()
plt.savefig('../visuals/category-wise')
```
